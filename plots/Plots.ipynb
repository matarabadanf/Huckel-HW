```python
from huckel import Chain
import matplotlib.pyplot as plt
```

```python
# Create a 10 atom chain
chain = Chain(10)
chain.plot_eigenvalues()

# Create a 10 atom ring
ring = Chain(10)
ring.close_ring()
ring.plot_eigenvalues()

plt.savefig('chain_vs_ring.jpg', dpi=800)
```

```python
# Create a 10 atom chain
chain = Chain(1000)
chain.plot_eigenvalues()

# Create a 10 atom ring
ring = Chain(1000)
ring.close_ring()
ring.plot_eigenvalues()

plt.savefig('convergence_limit.jpg', dpi=800)
```

```python
# Testing the influence alpha in the eigenvalues
chain = Chain(100)
chain.plot_eigenvalues()

chain_alpha_prime = Chain(100)
chain_alpha_prime.set_alpha_prime(1)
chain_alpha_prime.plot_eigenvalues(comment=r'$\alpha=1$')

chain_alpha_prime = Chain(100)
chain_alpha_prime.set_alpha_prime(2)
chain_alpha_prime.plot_eigenvalues(comment=r'$\alpha=2$')

plt.savefig('alpha_in_chain.jpg', dpi=600)
```

```python
# Testing the influence of beta in the eigenvalues
chain = Chain(100)
chain.plot_eigenvalues()

chain_beta_minus = Chain(100)
chain_beta_minus.set_beta_minus(-1.5)
chain_beta_minus.plot_eigenvalues(comment=r'$\beta=-1.5$')

chain_beta_minus = Chain(100)
chain_beta_minus.set_beta_minus(-.5)
chain_beta_minus.plot_eigenvalues(comment=r'$\beta=-0.5$')

plt.savefig('chain_beta_eigenvaules.jpg', dpi=600)
```

```python
# Testing the influence alpha in the eigenvalues for a ring
ring = Chain(100)
ring.close_ring()
ring.plot_eigenvalues()

ring_alpha_prime = Chain(100)
ring_alpha_prime.close_ring()
ring_alpha_prime.set_alpha_prime(1)
ring_alpha_prime.plot_eigenvalues(comment=r'$\alpha=1$')

ring_alpha_prime = Chain(100)
ring_alpha_prime.close_ring()
ring_alpha_prime.set_alpha_prime(2)
ring_alpha_prime.plot_eigenvalues(comment=r'$\alpha=2$')

plt.savefig('alpha_in_ring.jpg', dpi=600)
```

```python
# Testing the influence of beta in the eigenvalues for a ring
ring = Chain(100)
ring.close_ring()
ring.plot_eigenvalues()

ring_beta_minus = Chain(100)
ring_beta_minus.close_ring()
ring_beta_minus.set_beta_minus(-1.5)
ring_beta_minus.plot_eigenvalues(comment=r'$\beta=-0.8$')

ring_beta_minus = Chain(100)
ring_beta_minus.close_ring()
ring_beta_minus.set_beta_minus(-.5)
ring_beta_minus.plot_eigenvalues(comment=r'$\beta=-0.5$')

plt.savefig('ring_beta_eigenvalues.jpg', dpi=600)
```
