```python
from huckel import Chain
import matplotlib.pyplot as plt
```

```python
# Create a 10 atom chain
chain = Chain(10)
chain.plot_eigenvalues()

# Create a 10 atom ring
ring = Chain(10)
ring.close_ring()
ring.plot_eigenvalues()
```
