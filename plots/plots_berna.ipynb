```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the text file
datafiles = ['CV.txt']


for data in datafiles:
    data = pd.read_csv('CV.txt', delimiter=';')
    data['WE(1).Current (A)'] = data['WE(1).Current (A)']*1000/(np.pi * 0.15 **2)
    data = data[data['Scan'] == 3]
    plt.plot(data['Potential applied (V)'], data['WE(1).Current (A)'])

    plt.title('Densidad de corriente frente a potencial')
    plt.xlabel('Potencial aplicado ($V$)')
    plt.ylabel('Densidad de corriente ($A/m^2$)')
```
